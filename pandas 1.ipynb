```python
import pandas as pd
```

```python
a = pd.Series([45,23,65,35,98])
print(a)
print(type(a))
```

```python
a = pd.Series([45,23,65,35,98], index=[1,2,3,4,5], dtype="float")
print(a)
```

```python
a = pd.Series([45,23,65,35,98], index=['a','b','c','d','e'])
print(a)
```

```python
a = pd.Series([45,23,65,0,98], dtype="bool")
print(a)
```

```python
a = ['Maths', 'chemistry', "Physics"]
b = [28, 98, 76]
c = pd.Series(b, index=a)
print(c)
```

```python
a = pd.Series([45,23,65,35,98], index=['a','b','c','d','e'], name='Pandas Series')
print(a)
```

```python
a = pd.Series({'hardik':55,'bhumik':55,'rutwik':55})
print(a)
```

```python
x = []
for var in range(1, 3001):
    x.append(var)
```

```python
a = pd.Series(x)
print(a)
```

```python
df = pd.DataFrame({'maths':[28,25,29],'science':[34,45,76],'english':[45,35,76]}, index=["Priyraj", "Krish", "Dhyey"])
print(df)
```

```python
frame = pd.DataFrame([{'a':10,'b':20,'c':30},{'b':10,'a':20,'c':15}])
print(frame)
```

```python
i = ['Maggi', 'pasta', 'wafers', 'soap']
df = pd.DataFrame({"wholesale":[18,24,8,28],"Selling":[20,25,10,30]}, index=i)
print(df)
```
